Fonction pour recuperer tous les departements

```python
def extract_values_from_json(filepath, key_to_extract):
    """
    Charge un fichier JSON et extrait toutes les valeurs pour une clé specifiée dans une liste

    Args:
        filepath (str): chemin vers le fichier
        key_to_extract (str): la clé pour laquelle les valeurs doivent etre extraites

    Returns:
        list: Une liste contenant toutes les valeurs extraites pour la clé specifiée
              Vide si pas de fichier, fichier vide, pas de clé
    """
    extracted_values = []
    
    if not os.path.exists(filepath):
        print(f"Erreur : Le fichier '{filepath}' n'a pas été trouvé.")
        return []

    try:
        with open(filepath, 'r', encoding='utf-8') as f:
            data = json.load(f)
    except json.JSONDecodeError as e:
        print(f"Erreur : Impossible de décoder le JSON du fichier '{filepath}'. Le fichier est peut-être vide ou mal formé. Détails : {e}")
        return []
    except Exception as e:
        print(f"Une erreur inattendue est survenue lors de la lecture du fichier '{filepath}'. Détails : {e}")
        return []
    
    items_to_process = []
    if isinstance(data, dict):
        if "code" in data and isinstance(data["code"], list):
            items_to_process = data["code"]
        else:
            # Fallback: chercher la première liste trouvée parmi les valeurs de niveau supérieur
            for value in data.values():
                if isinstance(value, list):
                    items_to_process = value
                    break
    elif isinstance(data, list):
        # Si le JSON est directement une liste de dictionnaires
        items_to_process = data
    
    if not items_to_process:
        print(f"Attention : La structure JSON du fichier '{filepath}' ne contient pas de liste d'éléments à traiter sous des clés attendues (comme 'territoires' ou 'content') ou directement. Impossible d'extraire les valeurs.")
        return []

    for item in items_to_process:
        if isinstance(item, dict) and key_to_extract in item:
           
                extracted_values.append(item[key_to_extract])

    return extracted_values
```

Requete Offres d'emploi

```python
import requests

url = "https://api.francetravail.io/partenaire/offresdemploi/v2/offres/search"

headers_api = {
    'Authorization': f'Bearer {token}',
    'Accept': 'application/json',
}

params = {
    "grandDomaine": "M14",
    "departement":""

}

start = 0
step = 149  # nombre d'offres par requête
all_offres = []

for dep in departements:
    time.sleep(0.1)
    params["departement"] = dep

    while True:
        headers_api['Range'] = f"{start}-{start + step - 1}"
        response = requests.get(url, headers=headers_api, params=params)
        if response.status_code in [200, 206]:
            data = response.json()
            all_offres.extend(data.get("resultats", []))
            
            content_range = response.headers.get("Content-Range")
            if content_range:
                # Exemple de content_range : "offres 0-49/237"
                total = int(content_range.split('/')[-1])
                start += step
                if start >= total:
                    break
            else:
                break
        else:
            print(f"Erreur {response.status_code} : {response.text}")
            break

print(f"Nombre total d'offres récupérées : {len(all_offres)}")

```

Export JSON / Création DataFrame

```python
import json

# Sauvegarder toutes les offres dans un fichier JSON complet
with open('offre.json', 'w', encoding='utf-8') as f:
    json.dump({"resultats": all_offres}, f, ensure_ascii=False, indent=4)


with open("offre.json", "r", encoding="utf-8") as f:
    data = json.load(f)
```

```python
import pandas as pd 

# Transformer en DataFrame
records = []
for offre in data.get("resultats", []):
    record = {
        "id": offre.get("id"),
        "intitule": offre.get("intitule"),
        "description": offre.get("description"),
        "dateCreation": offre.get("dateCreation"),
        "dateActualisation": offre.get("dateActualisation"),
        "lieuTravail_libelle": offre.get("lieuTravail", {}).get("libelle"),
        "lieuTravail_codePostal": offre.get("lieuTravail", {}).get("codePostal"),
        "lieuTravail_commune": offre.get("lieuTravail", {}).get("commune"),
        "romeCode": offre.get("romeCode"),
        "romeLibelle": offre.get("romeLibelle"),
        "entreprise_nom": offre.get("entreprise", {}).get("nom"),
        "typeContrat": offre.get("typeContrat"),
        "typeContratLibelle": offre.get("typeContratLibelle"),
        "experienceExige": offre.get("experienceExige"),
        "experienceLibelle": offre.get("experienceLibelle"),
        "salaire_libelle": offre.get("salaire", {}).get("libelle"),
        "nombrePostes": offre.get("nombrePostes"),
        "contact_nom": offre.get("contact", {}).get("nom"),
        "contact_coordonnees": offre.get("contact", {}).get("coordonnees1"),
        "urlPostulation": offre.get("contact", {}).get("urlPostulation") or offre.get("contact", {}).get("coordonnees1"),
    }
    records.append(record)

df_offres = pd.DataFrame(records)

df_offres.head(100)
```

Ajout des skills par offres - SQL / PYTHON / POWER BI / SPARK / AZURE / EXCEL / DAX

```python
df_offres['SQL'] = df_offres['description'].str.contains('SQL', case=False, na=False)
df_offres['PYTHON'] = df_offres['description'].str.contains('PYTHON', case=False, na=False)
df_offres['POWER_BI'] = df_offres['description'].str.contains('POWER BI', case=False, na=False)
df_offres['SPARK'] = df_offres['description'].str.contains('SPARK', case=False, na=False)
df_offres['AZURE'] = df_offres['description'].str.contains('AZURE', case=False, na=False)
df_offres['EXCEL'] = df_offres['description'].str.contains('EXCEL', case=False, na=False)
df_offres['DAX'] = df_offres['description'].str.contains('DAX', case=False, na=False)
df_offres['AWS'] = df_offres['description'].str.contains('AWS', case=False, na=False)
df_offres['JAVA'] = df_offres['description'].str.contains('JAVA', case=False, na=False)
df_offres['SNOWFLAKE'] = df_offres['description'].str.contains('SNOWFLAKE', case=False, na=False)
df_offres['DATABRICKS'] = df_offres['description'].str.contains('DATABRICKS', case=False, na=False)


cols = ['SQL', 'PYTHON', 'POWER_BI', 'SPARK', 'AZURE', 'EXCEL', 'DAX', 'AWS', 'JAVA', 'SNOWFLAKE', 'DATABRICKS']

for col in cols:
    df_offres[col] = df_offres[col].astype(int)

df_offres
```

```python
cols = ['SQL', 'PYTHON', 'POWER_BI', 'SPARK', 'AZURE', 'EXCEL', 'DAX', 'AWS', 'JAVA', 'SNOWFLAKE', 'DATABRICKS']

# Créer une colonne avec la liste des compétences présentes pour chaque offre
def concat_skills(row):
    return ', '.join([col for col in cols if row[col] == 1])

df_offres['skills_list'] = df_offres.apply(concat_skills, axis=1)
df_offres
```
